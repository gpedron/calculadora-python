```python
# Identificar o usuário e Definir qual a operação matemática

nome = input("Escreva seu nome: ")

while True: # while para definir as operações e encerrar a calculadora( irá repetir até a escolha número 5)

    print("\n") #\n é um espaço entre as linhas
    print("Olá " + nome, "Antes de iniciarmos, escolha uma das operações matemáticas:")
    print("1 - Soma")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")
    print("5 - Sair")
    print("\n") #\n é um espaço entre as linhas

    opção = int(input("Escolha uma opção: "))

    if opção == 5:
        print("Calculadora encerrada.")
        break

    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))

    if(opção == 1):
        print("Soma: ", num1 + num2)

    if(opção == 2):
        print("Subtração: ", num1 - num2)

    if(opção == 3):
        print("Multiplicação: ", num1 * num2)

    if(opção == 4):
      if(num2 == 0):
        print("Não é possível dividir por zero.")
      else:
        print("Divisão:", num1 / num2)

    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")
```
